```python
def get_revenue_movie():
    revenue_movies = []
    tb = soup.find_all('td', class_='mojo-estimatable')
    # print(len(tb))
    for movie in tb[1::2]:
        revenue_movies.append(movie.text)
    return revenue_movies
```

```python
def get_titles_yearly():
    tb = soup.find_all('td', class_='a-text-left mojo-field-type-release mojo-cell-wide')
    # print(tb`)
    title_movies = []
    for movie in tb:
        title_movies.append(movie.select('a')[0].string)
    return title_movies
```
